```python
import matplotlib.pylab as plt
def plti(im, h=8, **kwargs):
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize = (w,h))
    plt.imshow(im, interpolation = "none", **kwargs)
    plt.axis("on")
%matplotlib inline 
im = plt.imread("plane.jpg")
print (im.shape)
im = im[:655,:1000,:]
#im * 25
plti(im)
```
